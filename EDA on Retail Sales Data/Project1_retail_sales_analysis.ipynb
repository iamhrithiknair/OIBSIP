```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\hrith\Downloads\Oasis\retail_sales_dataset.csv'
df = pd.read_csv(file_path)

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['YearMonth'] = df['Date'].dt.to_period('M')

# Gender-based analysis: Monthly total sales by gender
gender_sales = df.groupby(['YearMonth', 'Gender']).agg({'Total Amount': 'sum'}).reset_index()

# Convert 'YearMonth' to string for plotting
gender_sales['YearMonth'] = gender_sales['YearMonth'].astype(str)

# Plot gender-wise sales trends
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearMonth', y='Total Amount', hue='Gender', data=gender_sales, marker='o')
plt.title('Monthly Sales Trends by Gender')
plt.xticks(rotation=45)
plt.ylabel('Total Sales Amount')
plt.xlabel('Month')
plt.tight_layout()
plt.show()

# Age-based analysis: Monthly total sales by age group
# Create age bins
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Group data by Age Group and YearMonth
age_sales = df.groupby(['YearMonth', 'Age Group']).agg({'Total Amount': 'sum'}).reset_index()

# Convert 'YearMonth' to string for plotting
age_sales['YearMonth'] = age_sales['YearMonth'].astype(str)

# Plot age group-wise sales trends
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearMonth', y='Total Amount', hue='Age Group', data=age_sales, marker='o')
plt.title('Monthly Sales Trends by Age Group')
plt.xticks(rotation=45)
plt.ylabel('Total Sales Amount')
plt.xlabel('Month')
plt.tight_layout()
plt.show()
```

```python
# Pivot table for heatmap
gender_category_sales = df.groupby(['YearMonth', 'Gender', 'Product Category']).agg({'Total Amount': 'sum'}).reset_index()

gender_category_pivot = gender_category_sales.pivot_table(index='Product Category', columns='YearMonth', values='Total Amount', aggfunc='sum')

# Plot heatmap for gender-based product category sales
plt.figure(figsize=(12, 8))
sns.heatmap(gender_category_pivot, cmap="YlGnBu", annot=True, fmt='.0f')
plt.title('Monthly Product Category Sales by Gender (Heatmap)')
plt.xlabel('Month')
plt.ylabel('Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Pivot table for heatmap
age_category_sales = df.groupby(['YearMonth', 'Age Group', 'Product Category']).agg({'Total Amount': 'sum'}).reset_index()

age_category_pivot = age_category_sales.pivot_table(index='Product Category', columns='YearMonth', values='Total Amount', aggfunc='sum')

# Plot heatmap for age-based product category sales
plt.figure(figsize=(12, 8))
sns.heatmap(age_category_pivot, cmap="OrRd", annot=True, fmt='.0f')
plt.title('Monthly Product Category Sales by Age Group (Heatmap)')
plt.xlabel('Month')
plt.ylabel('Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
